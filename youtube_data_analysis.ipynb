## Carga de datos

```python
import pandas as pd
us_videos_df = pd.read_csv('USvideos.csv')
us_videos_df
```

```python
us_category_df = pd.read_json('US_category_id.json')
us_category_df
```

# Reconocimiento de datos
### Descripción de los datos USvideos.csv
* video_id: Identificador único del video
* trending_date: Fecha en la que el video se volvió tendencia YY.DD.MM
* title: Título del video
* channel_title: Nombre del canal
* category_id: Categoría del video
* publish_time: Fecha de publicación del video YY-MM-DDTHH:MM:SS.MMMZ
* tags: Etiquetas del video, separadas por |
* views: Número de vistas
* likes: Número de likes
* dislikes: Número de dislikes
* comment_count: Número de comentarios
* thumbnail_link: Enlace a la miniatura del video
* comments_disabled: Indica si los comentarios están deshabilitados
* ratings_disabled: Indica si las calificaciones están deshabilitadas
* video_error_or_removed: Indica si el video fue eliminado o tiene errores
* description: Descripción del video
### Descripción de los datos US_category_id.json
* kind: Tipo de objeto
* etag: Etag del objeto
* items: Lista de categorías
    * kind: Tipo de objeto
    * etag: Etag del objeto
    * id: Identificador de la categoría
    * snippet: Detalles de la categoría
        * channel_id: Identificador del canal
        * title: Título de la categoría
        * assignable: Indica si la categoría es asignable

# Pre procesamiento de datos

## Selección de datos
Se eliminaran las columnas que no son necesarias para el análisis, para reducir la cantidad de datos y mejorar la eficiencia del análisis

```python
us_videos_df = us_videos_df.drop(['thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed', 'video_id'], axis=1)
us_videos_df.head()
```

```python
us_category_df = us_category_df.drop(['kind', 'etag'], axis=1)
us_category_df.head()
```

## Limpieza de datos
Se eliminarán los datos duplicados y los datos nulos

```python
us_videos_df = us_videos_df.drop_duplicates()
us_videos_df = us_videos_df.dropna()
```

## Cambio de tipos de datos

```python
us_videos_df.dtypes
```

```python
# Se debe cambiar el tipo de dato de trending_date y publish_time a datetime
us_videos_df['trending_date'] = pd.to_datetime(us_videos_df['trending_date'], format='%y.%d.%m')
us_videos_df['publish_time'] = pd.to_datetime(us_videos_df['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
us_videos_df.dtypes
```

```python
us_videos_df.head()
```

# Preparación de los datos
Se van a combinar los dos dataframes para poder comprendender mejor los datos
Se obtendrá el nombre de cada categoría y se añadirá al dataframe de videos

```python
category_name_df = pd.DataFrame(us_category_df['items'].apply(lambda x: x['snippet']['title']))
# Se cambia el nombre de la columna
category_name_df.columns = ['category_name']
category_name_df['category_id'] = us_category_df['items'].apply(lambda x: x['id'])
# Se cambia el tipo de dato de category_id a int
category_name_df['category_id'] = category_name_df['category_id'].astype(int)
category_name_df.head()
```

```python
us_videos_df = us_videos_df.merge(category_name_df, on='category_id')
us_videos_df.drop('category_id', axis=1, inplace=True)
us_videos_df.head()
```
